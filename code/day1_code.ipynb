```python
import data_parsing 

input_data = '../data/day1_data.txt'

twisty_turns = data_parsing.read_lines_of_text(input_data)

def return_next_step(current_step: int, rot_direction: str, rot_steps: int):

    # it's possible to loop around multiple times, so use modulo to get remainders beyond mutliples of 100
    if rot_direction == 'R':
        next_pos = current_step + (rot_steps % 100)
    if rot_direction == 'L':
        next_pos = current_step - (rot_steps % 100)

    # allow for negatives and fix that offset
    # the modulo term above should get rid of any chance of additional wrap arounds
    if next_pos < 0:
        next_pos = next_pos + 100
    
    if next_pos % 100 > 0:
        next_pos = next_pos % 100
    
    if next_pos % 100 == 0:
        next_pos = 0

    return next_pos

def count_zerovisits_and_give_newvalue(current_step: int, rot_direction:str, rot_steps:int):
    passthru = 0

    # count the number of  wrap arounds inherent in number of steps
    if int(abs(rot_steps)/ 100) > 0:
        passthru = int(abs(rot_steps)/100)

    # now calculate the extra passthrus, when it's not like you're wildly spinning the dial
    # use to our advantage that we already know the current step btwn 0 and and 99
 
    next_pos = return_next_step(current_step, rot_direction, rot_steps % 100)

    # if next position is 100 then it resets to 0, so that counts as passing thru/landing on zero
    if (next_pos == 0):
        passthru +=1    
    if ((next_pos != 0) & (current_step != 0) & (next_pos < current_step) & (rot_direction == 'R')):
        passthru +=1
    if ((next_pos != 0) & (current_step != 0) & (next_pos > current_step) & (rot_direction == 'L')):
        passthru += 1

    return passthru,next_pos
    

def do_day1_part1(user_data, initial_value):
    stop_points = []

    next_point = int(initial_value)
    for line in user_data:
        # only pass lines that have data in the expected format
        # where expected is first character denotes right or left
        # and make sure there is a numeric value for clicks
        allowed_rotations = ['L', 'R']
        if ((line[0] in allowed_rotations) & (line[1:].isnumeric())):
            direction = line[0]
            distance = int(line[1:])
            next_point = return_next_step(next_point, direction, distance)
            stop_points.append(next_point)
    return stop_points.count(0)


def do_day1_part2(user_data, initial_value): 
    zero_pass = []

    next_point = int(initial_value)
    for line in user_data:
        # only pass lines that have data in the expected format
        # where expected is first character denotes right or left
        # and make sure there is a numeric value for clicks
        allowed_rotations = ['L', 'R']
        if ((line[0] in allowed_rotations) & (line[1:].isnumeric())):
            direction = line[0]
            distance = int(line[1:])
            zero_counts,next_point = count_zerovisits_and_give_newvalue(next_point, direction, distance)
            zero_pass.append(zero_counts)
    return sum(zero_pass)



start_pos = 50

print ('Part 1: ', do_day1_part1(twisty_turns, start_pos))
print ('Part 2: ', do_day1_part2(twisty_turns, start_pos))

     
```

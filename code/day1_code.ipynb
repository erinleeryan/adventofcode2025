```python
import data_parsing 

input_data = '../data/day1_data.txt'

twisty_turns = data_parsing.read_lines_of_text(input_data)

def return_next_step(current_step, direction):
    ## to add later: test that current step is numeric
    ## test that direction starts with either L or R and subsequent values are numeric

    current_step = int(current_step)

    # split the turn into the direction and the number you step in that direction
    dial_direction = str(direction[0])
    step = int(direction[1:])

    if dial_direction == 'R':
        next_pos = current_step + step
    if dial_direction == 'L':
        next_pos = current_step - step

    # allow for negative wrap arounds
    if ((next_pos % 100 < 0) | (next_pos < 0)):
        next_pos = next_pos + 100
    
    if next_pos % 100 > 0:
        next_pos = next_pos % 100
    
    if next_pos % 100 == 0:
        next_pos = 0

    return next_pos

def count_zero_passthrus(current_step, direction):
    ## to add later: test that current step is numeric
    ## test that direction starts with either L or R and subsequent values are numeric

    current_step = int(current_step)

    # split the turn into the direction and the number you step in that direction
    dial_direction = str(direction[0])
    step = int(direction[1:])

    passthru = 0

    # count the number of  wrap arounds inherent in number of steps
    if int(abs(step)/ 100) > 0:
        passthru = int(abs(step)/100)

    # now calculate the extra passthrus, when it's not like you're wildly spinning the dial

    # use to our advantage that we already know the current step btwn 0 and and 99


    if dial_direction == 'R':
        next_pos = current_step + (step - passthru*100)
    if dial_direction == 'L':
        next_pos = current_step - (step - passthru*100)

    
    # if next position is 100 then it resets to 0, so that counts as passing thru/landing on zero
    if (next_pos == 0):
        passthru +=1    

    if ((next_pos < 0) & (current_step !=0)):
        passthru +=1
    if next_pos >= 100:
        passthru += 1

    #if (int(abs(step)/100) > 0):
    #    print (current_step, step, dial_direction, next_pos, passthru)

    return passthru
    


rest_pos = 50

pointers = []
zero_passes = []

for line in twisty_turns:
    #print (rest_pos)

    all_zero_pass = count_zero_passthrus(rest_pos, line)
    zero_passes.append(all_zero_pass)

    rest_pos = return_next_step(rest_pos, line)
    pointers.append(rest_pos)


#print (zero_passes)

print ('Part 1: ', pointers.count(0))
print ('Part 2:', sum(zero_passes))
     
```

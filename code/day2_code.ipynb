```python
import data_parsing 
import math

input_data = '../data/day2_data.txt'

sequences = data_parsing.read_a_csv(input_data)


### this first function is related to me not understanding the requirements at first
def has_digit_repeats(value: int):
    # set the repeats flag immediately
    repeats = False

    # because repeats flag already set to False 
    # it will automagically return a statement of False
    # if there's only a single digit in the passed value
    value_list = list(str(value))

    if len(value_list) > 1 :
        difference = []
        for x in range(1,len(value_list)):
            difference.append(int(value_list[x])-int(value_list[x-1]))

        if difference.count(0) >= 1:
            repeats = True

    return repeats

# this is what was really meant by the description of part 1
def comprised_of_two_repeating_sequences(value: int):
    value_list = list(str(value))

    if len(value_list) % 2 == 0:
        split_index = int(len(value_list)/2)
        first_half = ''.join(value_list[:split_index])
        second_half = ''.join(value_list[split_index:])

        if first_half == second_half:
            return value
        if first_half != second_half:
            return

def create_numerical_list(value_range:str):
    start,stop = value_range.split('-')
    num_list = list(range(int(start), int(stop)+1))
    return num_list
invalid_values = []

for sequence in sequences:
    num_range = create_numerical_list(sequence)


    for item in num_range:
        if (len(list(str(item))) % 2 == 0):
            if comprised_of_two_repeating_sequences(item) :
                invalid_values.append(item)

print ('Part 1: ', sum(invalid_values))

```

```python
# potential primes: find longest range in input
# divide length of value by 2
# generate range from 2 to length/2
# test for primes in that list
# make a generic primes list we just call from in there

# the sequences are the start, end, and a dash in the middle
# so if we subtract the one character for the dash divide by 2
# we'll know how many places exist in each number string
int_lengths = [int((len(x)-1)/2) for x in sequences]

# then to find the primes we need to create an
# array btwn of possible primes between 2 and the 
# number of places divided by two

integers = list(range(2,max(int_lengths)))
primes = [2]

for x in range(1,len(integers)):
    preceeding = integers[:x]
    divisible = False

    for test_case in preceeding:
        if integers[x] % test_case == 0:
            divisible = True

    if divisible == False:
        primes.append(integers[x])


def comprised_of_repeating_segments(value: int, num_segments:list):
    value_list = list(str(value))
    repeats = 0

    for num in num_segments:
        if len(value_list) % num == 0:
            split_index = int(len(value_list)/num)
            substrings = [] 
            for i in range(num):
                substrings.append(''.join(value_list[i*split_index:i*split_index+split_index]))
              
            #print (substrings)  
            if substrings.count(substrings[0]) == len(substrings):
                #print (substrings.count(substrings[0]), len(substrings))
                repeats +=1 

    if repeats == 0:
        return
    if repeats >= 1:
        return value


many_repeats = [] 
for sequence in sequences:
    num_range = create_numerical_list(sequence)
    for new_number in num_range:
        if comprised_of_repeating_segments(new_number, primes):
            many_repeats.append(new_number)


print ('Part 2 result:', sum(many_repeats))
```

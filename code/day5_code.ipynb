```python
import data_parsing
import numpy as np

input_data = '../data/day5_data.txt'
ingredients = data_parsing.read_lines_of_text(input_data)


break_line = [i for i, x in enumerate(ingredients) if x == '']

ingredients_to_check = ingredients[break_line[0]+1:]


# make a list of all the fresh ingredients

## way to fix things:
## sort by start and end, find overlaps and merge the
## ranges

fresh_ingredients = ingredients[:break_line[0]]

def reduce_size_fresh_ingredients(fresh_ingredient_list):
    holding_list = []

    for fresh in fresh_ingredient_list:
        start, end = fresh.strip().split('-')
        holding_list.append([int(start),int(end)])

    sorted_ingredients = sorted(holding_list)

    # because I sorted the holding list, notionally I don't need to edit the beginning
    # of any ranges, it should only by the end of a range

    for i,new_range in enumerate(sorted_ingredients):

        if (i == 0):
            reduced_list = np.array([new_range])

        if (i > 0):
            start = int(new_range[0])
            stop = int(new_range[1])
            lower_values = reduced_list[:,0]
            upper_values = reduced_list[:,1]

            replace_index = np.argwhere(((lower_values <= start) & (start <= upper_values)) | ((lower_values <= stop) & (stop <= upper_values)))

            # need where start or end is within a currently existing range
            # if start or end in the range
            # then modify the value in the holding list
            if len(replace_index) == 0:
                reduced_list = np.append(reduced_list, np.array([new_range]), axis=0)
            if len(replace_index) != 0:
                index = replace_index[0]
                needs_mod = reduced_list[index]
                if (needs_mod[:,0] > new_range[0]):
                    reduced_list[index, 0] = new_range[0]
                if (needs_mod[:,1] < new_range[1]):
                    reduced_list[index,1] = new_range[1]
    return reduced_list

fresh = reduce_size_fresh_ingredients(fresh_ingredients)
  
on_hand = 0

fresh_lower = fresh[0]
fresh_upper = fresh[1]



for ingredient in ingredients_to_check:
    check_index = np.argwhere((fresh[:,0] <= int(ingredient)) & (int(ingredient) <= fresh[:,1]))
    #print (ingredient, check_index)


    if len(check_index) != 0:
        on_hand += 1

print ('Part 1: ', on_hand)

fresh_ranges_to_counts = fresh[:,1]-fresh[:,0]+1
print ('Part 2: ', np.sum(fresh_ranges_to_counts) )

```

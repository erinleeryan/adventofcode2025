```python
import data_parsing 
import itertools

input_data = '../data/day3_example.txt'
battery_bank = data_parsing.read_lines_of_text(input_data)



def find_largest_combo_2digits(inputdigits:str):
    # first pass: find the max digit in the input
    # if it's not in the last place, then make a substring
    # of everything afterwards, find the max digit there
    # and return the combo

    # else: there is the possibility the max digit 
    # can be at the very end
    # so keep track of all digits that exist in the input
    # and after dropping the max digit, find the first position 
    # of the next largest
    input_split = [int(x) for x in inputdigits]
    max_digit = max(input_split)
    digit_range = list(range(0,10))
    digit_count = [input_split.count(x) for x in digit_range]

    # god this line feels janky when you're used to just
    # doing conditional slicing when using numpy
    digits_in_input = [val for i,val in enumerate(digit_range) if digit_count[i] != 0]
    digits_in_input.remove(max_digit)


    max_indices = [i for i, val in enumerate(input_split) if val == max_digit]
    if max_indices[0] < len(input_split)-1:
        end_string = input_split[max_indices[0]+1:]
        combo = int(str(max_digit)+str(max(end_string)))
    if max_indices[0] == len(input_split)-1:
        next_max_indices = [i for i, val in enumerate(input_split) if val == max(digits_in_input)]
        end_string = input_split[next_max_indices[0]+1:]
        combo = int(str(max(digits_in_input))+str(max(end_string)))

    return combo


battery_selections = []

for line in battery_bank:
    battery_selections.append(find_largest_combo_2digits(line))


print ('Part 1: ', sum(battery_selections))
```

```python
def find_battery_combos(inputdigits:str, num_batteries=12):
    # assume number of batteries is at least 2
    # first thing to do: subset the input so we're only 
    # looking at the numbers before those absolutely required for the battery solution

    # is this the coding equivalent of smacking the problem with a rock?
    # yes, this is incredibly effing inelegant. But it freaking works
    # Santa and the elves can put in a tech debt Jira ticket
    # to make this pretty and run faster and like most orgs
    # they'll ignore the tech debt ticket until it becomes a problem

    # note that if you give this num_batteries =2
    # it does properly return the same solution
    # as the part one solution up above....it just
    # probably takes longer because iterating between
    # combos is a poop show since there are so many

    input_split = [int(x) for x in inputdigits]

    prefix_list = input_split[:-(num_batteries-1)]
    start_digit = max(prefix_list)
    start_index = input_split.index(start_digit)

    continued_list = input_split[start_index+1:]
    solved_value = [start_digit] + input_split[start_index+1:start_index+num_batteries]
    solved_value = int(''.join([str(i) for i in solved_value]))

    for combo in itertools.combinations(continued_list, num_batteries-1):
        suggested_combo = list(combo)
        full_combo = [start_digit]+suggested_combo
        combo_int = int(''.join([str(i) for i in full_combo]))
        if combo_int > solved_value:
            solved_value = combo_int

    return solved_value


big_battery_set = []

for line in battery_bank:
    big_battery_set.append(find_battery_combos(line))

print( 'Part 2: ', sum(big_battery_set))
```

```python
def alt_find_battery_combos(inputdigits:str, num_batteries=12):
    
    input_split = [int(x) for x in inputdigits]
    print ('initial state: ', len(input_split))

    prefix_list = input_split[:-(num_batteries-1)]
    start_digit = max(prefix_list)
    start_index = input_split.index(start_digit)

    solution = [start_digit]
    remaining_list = input_split[start_index+1:]


    for i in range(num_batteries-1):
        remaining_batteries = num_batteries-1-i
        subset = remaining_list[:-remaining_batteries]
        next_digit = max(subset)
        solution.append(next_digit)
        remaining_list = remaining_list[remaining_list.index(next_digit)+1:]

    return int(''.join([str(i) for i in solution]))



big_battery_set = []

for line in battery_bank:
    big_battery_set.append(find_battery_combos(line, num_batteries=2))

print( 'Part 2: ', big_battery_set, sum(big_battery_set))




```
